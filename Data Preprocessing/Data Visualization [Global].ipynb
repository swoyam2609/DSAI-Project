```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Dataset/country.csv')
```

```python
df.head()
```

```python
df=df.transpose()
```

```python
df.head()
```

```python
df =df.set_axis(df.iloc[0], axis=1)
df =df[1:]
```

```python
df.head()
```

```python
df['Cases'] = df.sum(axis=1)
df = df[['Cases']]
df.head()

```

```python
df.to_csv("../Dataset/global_cases.csv", index=True, header=True, sep=",", mode="w")
```
